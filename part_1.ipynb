# Выпускной проект

**Банки — Сегментация пользователей по потреблению**

## Декомпозиция

**Основной заказчик:** Менеджер банка, который разрабатывает меры по удержанию клиентов.

**Основная цель:** Проанализировать клиентов регионального банка и выполнить сегментирование пользователей по количеству потребляемых продуктов и доле оттока клиентов. На основе выполненного анализа данных предложить на рассмотрение ответственного менедера предполагаемые меры по удержанию клиентов. 

**Основные гипотезы для рассмотрения:** 

Клиенты склонные к оттоку:
1. реже пользуются услугами банка (низкая активность).
2. имеют высокий или низкий баланс, что в одном случае свидетельствует о том, что клиент легко может перейти в другой банк на более привлекательных условиях, в другом случае, люди с низким балансом, могут отправится к конкурентам в поисках лучших кредитных условий.
3. не имеют кредитных краточек банка

**Основные этапы работы:**

1. **Предобработка данных:**
    1. Исследовать пропущенные значения и выполнить обработку пропусков;
    2. Исследовать соответствие типов данных и изменить типы данных (если потребуется);
    3. Проверить корректность наименований колонок и переименовать колонки (если потребуется);  
    4. Исследовать дубликаты;
    5. Удалить дубликаты (если есть);

2. **Исследовательский анализ данных:**
    1. Проверить влияние активности пользователей на вероятность ухода;
    2. Определить зависимость состояния баланса и вероятности ухода;
    3. Сопоставить количество потребительских продуктов (в том числе наличие кредитной карты) с вероятностью ухода;

3. **Для каждого клиента определить:**
    1. Количество используемых банковских продуктов;
    2. На основании полученных сведений разделить клиентов на 4 основных категории;
    3. Внутри каждой категории посчитать процент ушедщих;
    4. Внутри каждой категории определить дополнительные показатели (например, наличие кредитной карты, уровень лояльности, уровень зарплаты и показатель скоринга, возвраст, регион);
    5. Сделать выводы: как количество используемых банковских продуктов зависит от показателей клиентов;
    6. Сделать выводы: как количество используемых продуктов влияет на вероятность ухода.

4. **Подготовить материалы для составления презентации**

5. **Сформилировать общие выводы** на основе выделенных сегметнов. Для каждого выделенного сегмента сформилуровать **рекомендуемые меры** для удержания таких клиентов.
